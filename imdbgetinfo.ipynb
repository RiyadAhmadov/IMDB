```python
#get info from ibm movies name
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

url = "https://www.imdb.com/chart/top"
request = requests.get(url)
```

```python
my_soup = BeautifulSoup(request.content,'html.parser')
```

```python
#Tbody cekirik amma yalniz class-i lister-list olan tbody-ni
data = my_soup.find_all('tbody', {'class': 'lister-list'})

for tbody in data:
    rows = tbody.find_all("tr")
    for row in rows:
        print(row)
```

```python
data = my_soup.find_all('tbody', {'class': 'lister-list'})
l = []
for tbody in data:
    rows = tbody.find_all("tr")
    for row in rows:
        baslik = row.find_all('td',{'class':'titleColumn'})
        filmadi = baslik[0].text
        cleaned_item = re.sub(r'\s+', ' ', filmadi.strip())
        l.append(cleaned_item)
```

```python
print(l)
print(len(l))
```

```python
clean_name = []
clean_year = []
for name in l:
    trimmed_title = re.sub(r'[ \s*\(\d{4}\).]', '', name)
    trimmed_name = name[-5:-1]
    clean_name.append(trimmed_title)
    clean_year.append(trimmed_name)
clean_year
clean_name
info = pd.DataFrame(list(zip(clean_name,clean_year)), columns=['Name','Year'])
info
```
